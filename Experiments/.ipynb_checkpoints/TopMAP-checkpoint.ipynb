# Topologically Regularization of bifurcating trajectory

In this notebook, we show how a topological loss can be combined with a non-linear embedding procedure, as to regularize the embedding and better reflect the topological---in this case bifurcating---prior.

We start by setting the working directory and importing the necessary libraries.

```python
# Set working directory
import os
os.chdir("..")

# Handling arrays and data.frames
import pandas as pd 
import numpy as np

# Loading R objects into python
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri

# Functions for deep learning (Pytorch)
import torch
from torch import nn

# Pytorch compatible topology layer
from topologylayer.nn import AlphaLayer
from topembed import DiagramFeature

# Umap specific functions
from topembed import umap_initialize, umap_loss

# Plotting
import seaborn as sns
import matplotlib.pyplot as plt

# Ordinary PCA embedding
from sklearn.decomposition import PCA

%matplotlib inline
```

## Import data and view restriction to first two coordinates

```python
file_name = os.path.join("Data", "Toy", "noisy_Y.csv")
df = pd.read_csv(file_name, index_col=0)
milestone_id = df.loc[:,"b"].astype(str)
df = df.drop(["b"], axis=1)

sns.scatterplot(x=df.iloc[:,0], y=df.iloc[:,1], hue=milestone_id)
```

## Conduct ordinary UMAP embedding

```python
# UMAP parameters
n_iter = 1000
n_neighbors = 15
spread = 1.0
min_dist = 0.1
learning_rate = 5e-2

# Initialize UMAP embedding with PCA
P, Y_emb, a, b = umap_initialize(df, n_neighbors=n_neighbors, spread=spread, min_dist=min_dist)

# Conduct UMAP embedding
Y_emb = torch.autograd.Variable(Y_emb, requires_grad=True)
optimizer = torch.optim.Adam([Y_emb], lr=learning_rate)

for iter in range(n_iter):

    optimizer.zero_grad()
    loss = umap_loss(P, Y_emb, a, b)
    loss.backward()
    optimizer.step()
    Y_emb - torch.mean(Y_emb, 0)

    if iter == 0 or (iter + 1) % (int(n_iter) / 10) == 0:
        print ("[Iter %d] [embedding loss: %f]" % (iter + 1, loss.item()))        
        
Y_emb = Y_emb.detach().numpy()
sns.scatterplot(x=Y_emb[:,0], y=Y_emb[:,1], hue=milestone_id)
```

## Apply topological optimization to increase the number of leaves

```python
# Define topological optimization
def g(p): return p[1] - p[0]
top_layer = AlphaLayer(maxdim=0)
connected_feature = DiagramFeature(dim=0, j=3, g=g)
flare_feature = DiagramFeature(dim=0, j=3, g=g)
flare_threshold = 0.5
mu = 15

# Parameters for topological optimization
n_iter = 1000
learning_rate = 5e-2

# Conduct topological optimization       
Y_top = torch.autograd.Variable(torch.tensor(Y_emb), requires_grad=True)
optimizer = torch.optim.Adam([Y_top], lr=learning_rate)

for iter in range(n_iter):

    optimizer.zero_grad()
    f = torch.norm(Y_top - torch.mean(Y_top, dim=0), dim=1)
    f /= torch.max(f)
    dgminfo_connected = top_layer(Y_top)
    dgminfo_flare = top_layer(Y_top[f > flare_threshold,:])
    loss = (connected_feature(dgminfo_connected))**2 + (flare_feature(dgminfo_flare) - mu)**2
    loss.backward()
    optimizer.step()
    Y_top - torch.mean(Y_top, 0)

    if iter == 0 or (iter + 1) % (int(n_iter) / 10) == 0:
        print ("[Iter %d] [top loss: %f]" % (iter + 1, loss.item()))
        
Y_top = Y_top.detach().numpy()
sns.scatterplot(x=Y_top[:,0], y=Y_top[:,1], hue=milestone_id)
```

## Conduct topologically regularized UMAP

```python
# Define topological optimization
def g(p): return p[1] - p[0]
top_layer = AlphaLayer(maxdim=1)
connected_feature = DiagramFeature(dim=0, j=3, g=g)
flare_feature = DiagramFeature(dim=0, j=3, g=g)
flare_threshold = 0.5
mu = 15

# parameters for topologically regularized embedding
n_iter = 1000
n_neighbors = 15
spread = 1.0
min_dist = 0.1
learning_rate = 5e-2
lambda_top = 1e2

# Initialize topologically regularized embedding with PCA
P, Y_topemb, a, b = umap_initialize(df, n_neighbors=n_neighbors, spread=spread, min_dist=min_dist)

# Conduct topologically regularized embedding
df_loss = pd.DataFrame(index=np.arange(n_iter), columns=["umap", "top", "total"])
Y_topemb = torch.autograd.Variable(Y_topemb, requires_grad=True)
optimizer = torch.optim.Adam([Y_topemb], lr=learning_rate)

for iter in range(n_iter):

    optimizer.zero_grad()
    f = torch.norm(Y_topemb - torch.mean(Y_topemb, dim=0), dim=1)
    f /= torch.max(f)
    dgminfo_connected = top_layer(Y_topemb)
    dgminfo_flare = top_layer(Y_topemb[f > flare_threshold,:])
    topological_loss = lambda_top * ((connected_feature(dgminfo_connected))**2 + 
                                     (flare_feature(dgminfo_flare) - mu)**2)
    embedding_loss = umap_loss(P, Y_topemb, a, b) 
    loss = embedding_loss + topological_loss
    loss.backward()
    optimizer.step()
    Y_topemb - torch.mean(Y_topemb, 0)
    
    df_loss.loc[iter,:] = np.array([embedding_loss.item(), topological_loss.item(), loss.item()])

    if iter == 0 or (iter + 1) % (int(n_iter) / 10) == 0:
        print ("[Iter %d] [embedding loss: %f, topological loss: %f, total loss: %f]" % 
               (iter + 1, embedding_loss.item(), topological_loss.item(), loss.item()))

Y_topemb = Y_topemb.detach().numpy()
sns.scatterplot(x=Y_topemb[:,0], y=Y_topemb[:,1], hue=milestone_id)
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
sns.lineplot(ax=ax, data=df_loss.loc[np.arange(0, df_loss.shape[0], 3),:].astype(float))
ax.set_xlabel("iter", fontsize=15)
ax.set_ylabel("loss", fontsize=15)
```

```python
f = np.linalg.norm(Y_topemb - np.mean(Y_topemb, axis=0), axis=1)
f /= np.max(f)
sns.scatterplot(x=Y_topemb[:,0], y=Y_topemb[:,1], hue=f > flare_threshold)
```
